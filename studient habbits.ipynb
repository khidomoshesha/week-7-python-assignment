```python
# Student Habits and Performance Analysis
# Assignment: Data Exploration, Basic Analysis, and Visualization using pandas, matplotlib, and seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Apply seaborn theme to all plots
sns.set(style="whitegrid")

# Task 1: Load and Explore the Dataset

try:
    df = pd.read_csv("student_habits_performance.csv")
    print("‚úÖ Dataset loaded successfully.")
except FileNotFoundError:
    print("‚ùå File not found. Please make sure the file is in the correct directory.")

# Display first 5 rows
print("\nüìä First 5 rows of the dataset:")
print(df.head())

# Check for missing values
print("\nüîç Missing values in each column:")
print(df.isnull().sum())

# Drop rows with missing values
df_cleaned = df.dropna()

# Check data types
print("\nüîß Data types:")
print(df_cleaned.dtypes)

# Basic statistics
print("\nüìà Summary statistics:")
print(df_cleaned.describe())

# Task 2: Basic Data Analysis

# Average exam score by gender
grouped_gender = df_cleaned.groupby('gender')['exam_score'].mean()
print("\nüéì Average Exam Score by Gender:")
print(grouped_gender)

# Average study hours by part-time job
grouped_job = df_cleaned.groupby('part_time_job')['study_hours_per_day'].mean()
print("\nüíº Average Study Hours by Part-Time Job:")
print(grouped_job)

# Average exam score by diet quality
grouped_diet = df_cleaned.groupby('diet_quality')['exam_score'].mean()
print("\nüçè Average Exam Score by Diet Quality:")
print(grouped_diet)

# Task 3: Data Visualization

# 1. Line Chart ‚Äì Exam Score Trend (using index as proxy for time)
plt.figure(figsize=(10, 5))
sns.lineplot(x=df_cleaned.index, y=df_cleaned['exam_score'], color='blue')
plt.title("Exam Score Trend by Student Index")
plt.xlabel("Student Index")
plt.ylabel("Exam Score")
plt.tight_layout()
plt.show()

# 2. Bar Chart ‚Äì Average Exam Score by Diet Quality
plt.figure(figsize=(8, 5))
sns.barplot(x=grouped_diet.index, y=grouped_diet.values, palette='Greens_d')
plt.title("Average Exam Score by Diet Quality")
plt.xlabel("Diet Quality")
plt.ylabel("Average Exam Score")
plt.tight_layout()
plt.show()

# 3. Histogram ‚Äì Distribution of Study Hours
plt.figure(figsize=(8, 5))
sns.histplot(df_cleaned['study_hours_per_day'], bins=15, kde=True, color='orange')
plt.title("Distribution of Daily Study Hours")
plt.xlabel("Study Hours Per Day")
plt.ylabel("Number of Students")
plt.tight_layout()
plt.show()

# 4. Scatter Plot ‚Äì Study Hours vs. Exam Score
plt.figure(figsize=(8, 5))
sns.scatterplot(x='study_hours_per_day', y='exam_score', data=df_cleaned, color='purple')
plt.title("Study Hours vs. Exam Score")
plt.xlabel("Study Hours Per Day")
plt.ylabel("Exam Score")
plt.tight_layout()
plt.show()


```
